# 2020-03-09一切的一切，正在上演或者将要上演

原创 大文豪曼因斯坦  大文豪曼因斯坦  2020-03-09  

通过国内外的新闻想必很多人已经看到了正在发生的一切，经济也罢，动荡的政治格局也罢，该来的终究要来。

有些事最多只是迟到，绝不缺席。

很多人觉得我应该说些什么，在我看来此时早已无需多言。

很多人会说，如果不是疫情全球蔓延不会有如此局面，我要说的是即使没有疫情该来的依然会来，疫情只是相，绝非因。

该说的我早就在过去的文章和留言中早就说了，其中很多已经在一定程度上算是泄漏天机，当然如此也一定程度上与我有损。说了这么多我已是尽力而为，剩下的都是你的事情。我说过我救不了谁，只有你们自己才可以救自己。即便我能够拨开云雾看清未来，包括未来的危险与艰难，包括未来的希望与美好，可惜的是这些将与很多人无缘。

自从更新文章以来总共只有两篇文章因故未更，不是这两篇文章不重要，相反没有更新出来的这两篇反而更重要。

因为其中牵涉不可说不好说的内容，当然也牵涉了巨大的价值。

但是该提醒该警示的我相信我一点都没有落下。

这两篇因故未更的文章的名字相信老读者应该都知道，日后我也没有再发的打算。

之前总有人私信索要这两篇文章，从今天起禁止任何人私信索要这两篇文章。

我所说所断确实蕴含了巨大的价值，我一再奉劝各位不要以此谋利，切勿以此来投机，很多事不是你所能驾驭所能控制的。

做自己能够驾驭的事情，做属于自己该做的事情。

希望各位再这里看到的不是利，而是可以看到更为宽广更为真实的世界和未来。

根据能够看到的这些去努力做好自己该做的事情。

这个世界除了私利，除了贪嗔痴，还有爱，还有理想，还有情怀，还有责任和使命，也还有传承和价值体现，把眼界放宽一点，把视野放远一点，你的人生可能会更为精彩。

警醒之言无需再说，嘱咐之语也无需再道，有些话我说的很含蓄，但并不代表你可以很含蓄的去理解，甚至不理解。

实在听不懂的我也没辙，于初心而言我已尽力。还是那句话，来路艰难，珍重万千！

# 2020-02-24一盆冷水

原创 大文豪曼因  大文豪曼因斯坦  2月24日
之前推送的消息该提醒的也提醒了，该嘱咐的也嘱咐到了，近期本无意更新文章，但是看到有些人混淆是非，各种诱导大家实在是忍无可忍。

好吧，那我就给大家来浇一盆冷水。
 
## 央行有没有放水？
最近各种媒体自媒体等等到处鼓吹央妈又是放水又是降息的，动辄几千亿几万亿的放水说得跟真的似的，广大不明真相的群众更是奔走相告，仿佛自己的那点钱不抓紧投资升值放手里就马上要贬值一大截似的。

吓得我也赶紧把央行的数据拉出来一看究竟，我倒是要看看放了多少水啊，让大家如此兴奋，看完之后不得不一盆夹着冰块的冷水浇给大家，截止当前央妈2020年市场净投放资金为负的近6000亿。我就想问鼓吹放水的各位了，哪里放水了？不光没有放水而且实实在在的回笼资金近6000亿。

好吧，说完放水的事再来看看有没有降息吧，在春节之后的短期逆回购操作利率分别下降10个基点，一年期MLF操作利率也下调10个基点，MLF操作利率客观上也引导了LPR下行，但LPR下行却基本上不影响个人房贷利率，房住不炒仍然是当前房地产调控政策的主导方向，银行可通过LPR加点方式确定个人房贷利率，基本保持原有水平。那些还指望着我们再次拎房地产这把尿壶的朋友可以省省了，即便是新冠肺炎疫情如此严重，让全球最主要经济体几乎停摆一月以上的情况下，我们仍然毅然决然的坚持常态化货币政策，坚决不搞“大水漫灌”，绝不采取刺激宽松的货币政策。

我早就告诉过各位，没有援军了，该撤的几年前就撤了，你就好好站好最后一班岗吧，虽然有点高。

已经挺了几年了，也不差再多挺几年。

说实话我们货币政策空间充足么，工具箱足够么？

当然，可是即便如此我们现在仍然不用，为何。因为我们很清楚更难的还在后面，共克时艰的时候还没到呢，现在就把空间释放了，后面危机来临时难道你想我们像美元一样裸奔么？美联储加息提裤子都还没提到位，这不去年就迫不及待的又开始脱裤子了，所以在接下来的这轮危机中，美元即使不裸奔，那也差不了多少。
 
## 疫情影响知多少？
刚刚步入2020我们就遭遇了新冠肺炎疫情之变，而此次事件也让全球最重要的经济体几乎停摆一个月，综合考虑后续影响以及各产业恢复生成的状况来看，粗略估计对全年经济影响程度几乎接近于两个月经济停滞。而且从目前疫情防控状况来看，这个病毒难以对付，况且现在有全球蔓延之势，如果一旦蔓延开来其后果可想而知。
 
## 全球局势裂变在即
年前中美第一阶段的经贸协议真的破天荒的能够落地到文本签署，但这个协议谈成了多少，后面还有多少个阶段要谈？三个，还是三十个？不要忘了持久战谁才是的鼻祖。

说实话在我看来，中美通过一纸协议来解决两国在国际政治经济格局下的巨大争端是太过天真的。

太平洋很大，可是实在无法容得下中美两国超级大国。

基于此双方在经贸领域的摩擦，乃至升级出现在接下来的几个月之内一点都不奇怪。

你真的以为双方很在意第一阶段的经贸协议，如果你真这样认为只能说明你年轻太简单了。

眼下我们忙于应对疫情之变，但灯塔可一点都没闲着，在灯塔的威逼利诱之下北极熊正一步步滑向无可挽回的深渊。

而特大嘴也将于月末访问印度，不出意外的话印度很快将会被赶鸭子上架，南亚之变已经为期不远，而且此变极有可能是大变，克什么米尔这个火药桶如今已经被人掀开了盖子，全球局势的快速裂变怕是要就此揭幕。

七、教育公平，家国之情


中国群众，个个都是吃瓜无数，吃瓜必吐子儿。说句大实话，大家要求的无非就是教育公平。

多招收留学生，并非不可理解，也并非没有布局的价值。但是，不能不顾高等教育的基本规则、基本底线、基本体面，尤其是不能忽略教育公平、社会公平。葱大被撕，其实是近几年留学生和外来移民侵占中国人仅有的一些内部公共资源并持续走恶的真实反映。

在中国，教育不公平，已经是一个被讨论近20年的话题了。以前，人们聚焦的是城市教育与农村教育的不公平、重点学校教育与一般学校教育的不公平、学校教育与家庭教育的不公平、男童教育与女童教育的不公平、优等生教育与后进生教育的不公平、升学教育与就业教育的不公平、关注知识与关注能力的不公平。随着这些年贫富差距的加大，目前，日益焦灼的是一二线大城市与农村地区的教育不公平，以及重点学区与一般学区的教育不公平。如今，葱大三陪一的学伴制度，北大倾囊相授巨款给菲家学生却冷酷拒绝自家考生，这其中扯出的洋人贵国人贱的双标行为，陡然增加了一条更刺痛中国人心的教育不公平。中国的教育资源并不十分充足，虽然高等教育规模扩大了，高考升学率提高了，但是，优质高等教育资源依然稀缺。本来就僧多粥少，现如今居然把最好的那部分无偿奉送给本来不如自己的外人，这种怨愤不是出几个声明就能抚平的。

我们的教育早就进入到了市场经济的体系之内，不仅国家政府对教育有投入，个人家庭也投入的越来越多。除了房子，目前国人最关心的就是孩子的教育。而房子中价格最高的莫属学区房，学区房的价值最终在于送孩子进名牌大学。说到底，房子问题也还是教育问题。我们的教育政策滞后、资源严重不足、经济发展不平衡、地区差异巨大，受教育的机会和权利事实上是不公平也不平等的。富裕的家庭，可以在教育方面高昂的消费，为孩子买到受优质教育的机会和权利。而中低收入的家庭，无法靠自身的综合力量来让孩子受到好的教育。但是，这并不意味着他们没有要求教育公平的权利，政府对教育的各种投入和政策是社会资源的第二次分配，就是要调节和缩小这种不公平。各个地区和学校的教学设施、师资水平不一样，很大程度上导致高考的巨大差距，最后可以享受的高等教育资源不一样。而高等教育是大多数年轻人步入社会的敲门砖，关系到今后的生存权和发展权。这种竞争，受分配形式、国家投入、各地资源、教育政策的影响非常之大，可以说教育公平是社会公平的基础，高等教育的公平是教育公平的最终体现。

由于人口结构和社会阶层的不同，毋庸置疑，富裕人口是我们社会的一小部分。受教育的主体肯定是中低收入的家庭，而社会的主体更是这样。中国自古就重视教育，历代都提倡耕读传家。读书受教育是中低阶层向更高的阶层流动的重要途径，阶层流动是社会稳定和发展的基础。教育公平，人才辈出，经济才能有动力，社会才能安定，国家才能欣欣向荣，民族才有凝聚力，文化才能四海传播，文明才能延续。孔夫子早就点明：有国有家者，不患寡而患不均，不患贫而患不安。教育的不公平就是不均，而不均会引发社会的不安。人民是汪洋大海，载舟亦覆舟。

各个阶层的孩子通过十几年的寒窗苦读，步入了理想的大学校园，对未来满怀期望。但是，贵外贱内优待留学生的双标政策却给他们狠狠了浇了一盆冷水。在同一所大学里，学习生活各个方面的优质资源，居然还是留给了很多不如自己的外国留学生。这种现象还是校方积极倡导和迎逢的。

这盆非常没有民族自尊和文化自信的冷水浇灭了莘莘学子心中的热情，点燃的却是大众心头的怒火。


八、自尊自信，华夏有情


民族是文化的主体，文化是民族的灵魂。中华文明几千年，延绵不断，历朝历代以中央之国自居，集四海八荒之精华，从来不存在缺乏文化自信的问题。

但是，从1840年鸦片战争，到1937年卢沟桥事变，一百多年间，几代人的生生死死，无法言尽的苦难艰辛，不堪回首的屈辱折磨，江山民族风雨飘摇，文化传承受到重创。1949年，毛主席重整河山，中国几代人才，愚公移山的坚毅。那些年，我们成就过，自豪过，鼓舞过，欣慰过。几千年的辉煌，如梦如幻，国人的自负与不忿、焦躁和挣扎，异化成了一个民族心底隐隐作痛的自卑。1979年，小平改革开放，大千世界缤纷繁华深深震撼了封闭太久的我们。铆足劲头的追赶，住最豪的宅，开最快的车，品最醇的酒，追最美的人，穿最贵的衣，登最高的山，填最大的海，拍最稀世的珍宝，建最雄伟的桥梁。到如今，敢磨出最利的刀，挑战最狠的人。

可是，回过头来，在自己的下一代身上，居然还是搞着贵人贱己的双标政策。可见，物质的堆砌与赶超，没有彻底的抹掉整个民族心底的自卑与彷徨。一百多年的断代梦魇，巅峰跌落泥沼的悲凉不堪，深深烙印在民族的记忆。而西方工业革命之后的科技爆发，无孔不入的文化侵蚀，强势的文化殖民，方方面面的碾压，使我们相形见绌，让我们缺乏自信，把一些敏感脆弱的负面民族情绪变相成一种维护民族自尊心的可笑行为。

北大几十万高额奖学金给连汉语都说不好的菲家学生，葱大强促自己的姑娘小子去陪伴黑哥们儿，正是暴露了中国高等教育的不自信。这背后是民族自信和文化自信的不足，是所谓民族自尊心虚掩着的脆弱的自卑心。这种自卑、爱面子使得很多国人在对待洋人（无论对方是黑是白）的时候都很在乎人家对自己的看法。搞得自己在洋人面前变成了一种虚假的表演，一味的迎合克己，损己利人，只为人家能给个好评。

彼时，哥哥在小米穷乡僻野的雪村卖艺求学，系里最活跃的当家年少白俊男对中国loser的哥哥那是一百个看不上眼。彼此见面除了塑料性质的how r u，别无他言。某日，数据分析课上，我两雄性动物被迫相邻而坐。好死不死，介位自视甚高的主人翁囧在了“三七得几”这个难题上，挑衅的问哥哥。哥哥俺惜字如金：二十一。从此，小白兔情根深种，再也离不开渣男哥哥俺了。每次上课都早早给哥占好位子，每次实验都一人扛两台大本子。哥哥享受的心安理得。因为哥哥知道，这是自己凭实力撩得的滔滔江水、连绵不绝。

谁成想，回到寄几家，居然一堆杀马特这样谄媚骚首的对待外人。这样能够换来人家的尊重和敬佩么？殊不知，这个世界上没有傻瓜，白人不傻、老墨不笨、阿三狡猾，黑哥们脑子里更不是墨鱼汁。

曾几何时，我们把象牙塔里的莘莘学子称为天之骄子。培养天之骄子的地方，如此自卑自贱，天之骄子的傲娇从何而来，如何有勇气指点江山、激昂文字，怎么有胸怀去粪土当年万户侯？高校作为教书育人的大本营，首先应该教会学生自尊自信、不卑不亢，而不是用这种贵人贱己的方式，来打击自己少年的民族文化的自尊自信。国民的尊严与民族的自信不是靠优待外国人得来的，而是需要从自我尊重开始做起。

 

老大提出文化自信是一个国家、一个民族发展中更基本、更深沉、更持久的力量；要不断增强中华民族的归属感、认同感、尊严感、荣誉感。一个民族，文化必须自信，否则就会丧失立足的根本，丢失延续的基脉。少年强则国强。大学时代，是青少年三观人格最关键的养成时期，如何让青少年像老大说的那样有民族自信和文化自信？高等教育是国之重器，难道礼部没有认真思考如何为国家民族担负起培养自信自强的下一代这个问题么？从目前的情况看，起码东南西不大和他那些杀马特小兄弟们没有做到这一点。

有人会问，我们凭什么自信？为什么自信？

5000多年绵延不绝、灿烂辉煌的华夏文明，滋养着中华民族生生不息、发展壮大，也为人类文明做出了不可磨灭的重大贡献。新中国成立特别是改革开放以来，我们几代人呕心沥血，用几十年时间走过了发达国家几百年走过的历程，创造了人类社会发展史上惊天动地的奇迹。70年披荆斩棘，70年风雨兼程，华夏大地发生了翻天覆地的变化，无论在中华民族历史上，还是在世界历史上，都是一部感天动地的奋斗史诗。因为奋斗和成就，我们有实力自信。因为要有更大的成就，我们需要更加坚定的信心。

从更大的格局来看，剑桥学者马丁.雅克的观点非常准确的解答了中国人何以自信这个问题。马丁.雅克认为，现代国家的政治概念都是由西方确立，实际是以民族（nation）主义为核心的政治实体。欧洲的如德意志民族，法兰西民族，亚洲的如大和民族，朝鲜民族。尽管这些民族是不断演化，不断分裂的一个群体，但正是由于当前的民族认同，才形成了对目前国家的认同。而中国的国家认同完全不是一个概念，他是一个文明文化（civilization）的认同，汉族不是血缘族群意义上的民族，而是作为华夏文明的延续和捍卫者的文化族群。战国时代，三国时代，民族概念远没有形成的年代，正是对华夏文明圈的概念促使割据国家不断有统一的使命感。历史上有过很多帝国，奥斯曼帝国波斯帝国，罗马帝国。在帝国崩塌后，再没有国家以文明价值为旗帜，去实现统一帝国统一。而军事征服形成的帝国，总会消失而不复现。而中国古语有云“天下合久必分，分久必合”，则是对文化的高度自信。民族主义划分的国家，随着历史演进，只会越来越小，例如欧洲的日耳曼民族如今已经分裂成无数国家，而文明主义划分的国家，只要文明旗帜不倒，国家就不会消亡。即便是美国作为新兴文明主义国家，如果在当今强大的时候没有形成千年文明认同价值体系，在帝国崩塌之后，也将形成无数小国，而无法象中国一样有帝国再现之荣耀。

文化自信问题不仅属于文化，它与国家的强大、民族的独立密不可分。即便在上个世纪那段水深火热的年代，鲁迅先生也没有失去对中华民族的自信。他说过：我们自古以来，就有埋头苦干的人，有拼命硬干的人，有为民请命的人，有舍身求法的人……这就是中国的脊梁。鲁迅的这番话，无疑不是在强调，我们的民族，我们的文化在人血馒头的时代也并没失掉自信力。那么，在我们已经卧薪尝胆，顽强的重新站立起来的今天，有什么理由要失去自信和从容的心态呢？

一个民族有了自信的心态，才能够使自己的文化更加的灿烂并广泛传播。

最近，迪士尼拍摄了真人版电影，花木兰。这恰恰是一个让我们审视欣赏民族文化的契机。在木兰之前，我们耳熟能详的公主们大约是：吃了毒苹果被王子的吻拯救的白雪公主；因为水晶鞋被王子救出恶毒继母家中的灰姑娘；被王子的真爱之吻破除沉睡诅咒的睡美人。而木兰，不再依赖男性，而是能与其互帮互助、分庭抗礼，甚至不需要王子，也能活出自己的光彩。她不仅是14位迪士尼公主中唯一来自中国的公主，更是唯一的没有魔法、没有王冠和公主裙、更没有被王子拯救的公主。

人类文明几千年，各民族的文化多姿多彩，即便是工业革命以来强调独立自主的西方欧美文化，也没有沉淀出一位像木兰一样的不动如山、处事泰然、方寸不乱、遵守纪律的女性形象。在中华文明中，顶天立地的英雄们如群星灿烂。木兰虽然仅流传在一首小诗当中，但是，这就足以令我们整个民族为之自豪。我们的民族几千年传承繁衍，是因为我们坚韧的文化基因悄无声息的造就了代父从军的木兰，守住了我们民族生生不息的底线。一个拥有如此丰富文化遗产的中国，应该有的文化大国文化强国的样子。

正视自己，兼容并包，才是大国气象的文化自信。

今年3月，央视播出了一部纪录片《从<中国>到中国》。其中的<中国>源于1973年，意大利导演米开朗基罗.安东尼奥尼拍摄的记录片《CHUNG》。2004年之前，《CHUNG》在中国是一部禁片，遭到很多批判。虽然很多人，比如哥，已经从各种渠道一睹这部禁片的风采。但是，这次央视能够积极的接受并包容这部珍贵的影片，作为纯种中国人，哥哥真心感到我们的民族自信心更强了。当年，安东尼没有一味采用官方安排的摆拍，而是深入到人们的日常生活中记录珍贵的瞬间。所以，中国官方大力批判纪录片《CHUNG》，原因是：丑化中国人。但究其根本原因，还是当时的中国缺乏文化自信。40年过去了，中国正逐渐褪去自卑和青涩，可以自信的正视自己，平视他人。正因为这份自信，我们今天才能以淡定、从容的姿态，去回望，去审视，去面对这些历史镜头。正是这份自信，今天，我们才能通过安东尼奥尼的镜头，看到我们的父辈，曾经的祖国，感受到血脉的延绵，文化的传承。

建立一个文化繁荣兴盛的大国，其难度堪比建设一座精神的万里长城。最重要的就是这份经历苦难之后，沉淀出从容的气度。就像老大一再强调的，文化自信对民族复兴极端重要性，国无精神不强，人无精神不立。一个没有精神力量和缺乏文化自信的民族，犹如无本之木，难于屹立世界民族之林。


九、华夏复兴，不矫情、不欠情


中华民族之所以历经磨难而不屈，饱经风霜犹自立，主要就在于“培育和发展了独具特色、博大精深的中华文化，为中华民族克服困难、生生不息提供了强大精神支撑”。

随着近代国家和民族的衰落，先进的同胞们不断探索救亡图存道路，从器物、制度到思想文化层面，为民族复兴做了不同意义和程度上的努力，实际上开启了我们实现近代化（现代化）的历史进程。实现中华民族伟大复兴，就是中华民族近代以来最伟大的梦想。1840年之后，作为一个自觉而明确的民族梦想，孙中山先生在中国历史上第一个喊出“振兴中华”的时代强音，发出了谋求民族复兴的先声。大革命时期，李大钊提出中华民族“重复振兴”问题，实现党内“民族复兴”的破题。毛主席手持红缨之日，更是慨然宣示：我们中华民族有同自己的敌人血战到底的气概，有在自力更生的基础上光复旧物的决心，有自立于世界民族之林的能力。可以说，实现中华民族伟大复兴，贯串了中国历史和中国社会的不同时代、不同阶级、各种政权和党派，是近代以来中国人民奋斗征程的最大夙愿，是中华民族发展历程的总灵魂。时至今日，这个民族复兴的梦想，凝聚着一百多年，几代中国人的心中所愿，体现了中华民族的整体利益，是每一个中华儿女的共同期盼。

我们要重建具有世界领导水平、文明模范角色、历史进步意义的文化感染力和政治感召力，在此基础之上，重塑充分的国民幸福感、强大的国家总实力和高度的民族自信心，在世界范围重现独具魅力的“中国气象”。什么气象？也就是盛唐曾呈现出的那么一种雍容而洒脱的气度、磅礴而开放的气势、强大而亲和的气场、自信而浪漫的气质。

一九四九年十月一日，在新中国成立的宣言中，毛主席向1840年以来为国家独立民族解放事业献身的先烈致敬。这样的先烈有多少？约两千万！这两千万革命先烈和川普以后各行各业为民族复兴鞠躬尽瘁的先辈们，是我们的榜样，也给我们力量。如今，中华儿女依然千千万，依然重任在肩。一代人是一代人的肩膀，一代人有一代人的责任，文化的传承，民族的复兴，不是靠口号，而是靠实干。

我们有神威·太湖之光超级计算机，被称为国之重器；世界上最大的22000吨起重机龙门吊“宏海”号；世界最大的水轮发电机；世界最大的“天眼”望远镜；世界领先的全路面起重机；世界最大抢险救援机器人；世界最大模锻液压机。我们还有国之利器的国产航母，万吨大驱；保家卫国的歼-20，直-20，运-20，无人机。我们更有国产大飞机，东风快递，核潜艇，让敌人闻之丧胆。天宫嫦娥，九天揽月，南海种岛，五洋捉鳖。毫不矫情的说，我们做得到。

此外，对待非洲兄弟，一带一路，我们一点也不欠情。5月，哥哥外审了一篇肯尼亚留学生的硕士论文，主题是评估中国在肯尼亚的“技术职业教育与培训”项目。我们的帮助使当地人的职业素质和技能有了明显的提高，受益匪浅。类似这样的项目，我们在各个后进国家有很多很多，都提供了真正的帮助，让他们实实在在的受益。他们的学生到我们家来学习先进的知识技能，我们欢迎，也倾囊相授。中国人的骨子里，有一种所谓克己利人的文化基因，宁可人可负我，不可我负人。但是，双标待遇实在不必。我们不欠任何人的情，不要把好事做成不堪。所谓，“斗米养恩,担米养仇”。礼尚往来，真诚以待，是我们文化基因中的大气、包容和善良，但是决不能偏颇，而要有度。

今天，我们比历史上任何时期都更加接近、更有信心、更有能力实现中华民族伟大复兴的中国梦，我们应该怎么办？是自轻自贱自伤？自尊自信自强？这个问题的答案值得深思，更要在脚踏实地的行动中深刻的体现。

家国与共，无论哪种选择，具体的行动，时时刻刻在进行。入了8月，全球躁动就像帝都的天气，酷热而又阴雨连绵。各种路边社消息层出不断。港村骚乱，愈演愈烈；弯弯待选，个个想弯不想直；三胖那边不明飞行物（注意，不是UFO）好欢脱；小米棒子又军演；小米家因为黑绿混色还枪击；欧诸已经糊成一锅粥；毛熊悄悄给俺们交家伙什儿；中国和米国的毛衣占，线头始终拉扯不断。川普三威四逼，又降息又3000亿，我们还是坚守底线。哥哥寂寞难耐，都是道上的，哥几个能不能来点儿实际的？果然，红豆终于破7，没在怕的。之前，中国内，很有一部分人，慌慌张张，唯恐红豆破7。殊不知，小米才是最怕的。股市是小米的命根子，看看股指：标普500指数收跌87.31点，跌幅2.98%，报2844.74点。道琼斯工业平均指数收跌767.27点，跌幅2.90%，报25717.74点，盘中一度暴跌将近1000点。纳斯达克综合指数收跌278.03点，跌幅3.47%，报7726.04点。科技股下跌，苹果大跌3%。按美股来说，这是赤果果的暴跌了。川普虽然心急，但还是贴心的给我们一个小瓜解暑：原来一直说中国干预汇率，所以是汇率操纵国。现如今，我们放弃干预，让红豆破7。川普却真的把我们定性为“汇率操纵国”了，原因竟然是：中国停止了对汇率的干预。哥哥简直要觉得川普同志可爱了。

我们就是铮铮铁骨有底气，我们就要乱世之中主沉浮。

新时代，中华民族伟大复兴的时代，我们要不卑不亢的大声说一句：不是我们要求洋人，而是洋人该求我们。这是我们自己赢得的，也是应得的。


十、血脉传承，真情所系


风雨砥砺不忘初心，春华秋实继往开来。我们忍辱负重，我们苦干艰辛，我们渴望复兴。为了什么？为了我们的民族血脉延续，为了我们的文明传承广播。

毫不夸张，毫不自大的说，中华民族的基因是世界上最优秀的基因。要保持这份基因的纯洁和延续，我们必须要做到：文化血脉不可断，民族基因不可改。

最近港村闹得欢，为什么这些跳梁小丑会沦落如此？因为港灿对自己的民族没有归属感，对自己的基因没有认同感。港灿灵魂中被殖民者深深烙印的奴性，使他们的精神失去了传承，疯狂的挣脱母亲的怀抱，一心要把自己变成了另一个亚细亚孤儿。

港村回归已有二十二年。如今20岁左右的一代，从未亲身经历过港英的殖民时代；30岁左右的一代，港英时代只是婴幼儿时候纸尿布的记忆；40岁左右的一代，殖民统治是他们的青春叛逆期；50左右的一代，如果有深刻的记忆，难道觉得那是他们人生的辉煌时刻？难道不记得：英属殖民的香港，黄皮肤的港村人连半分的民主和自由都没有，是彻彻底底的殖民地三等人待遇，所有的政府机构，遍布黄头发蓝眼睛的英国人还有黑头发褐皮肤的印度人。在如今有自由选举权，还有平等盛世的当下。港灿身归心不归，居然难忘殖民时期作为下等人的殊荣，对自己的民族没有丝毫认同感，看着血缘上同文同种的的兄弟们突然都茁壮成长，发展的比自己更好更强大时，嫉妒与怨恨就难以抑制。

香港从一个小渔村，能够一时兴起，不过是借他人之力。港灿们不愿面对这个现实。迁怒于回归。于是，不甘演化成了不忿，不忿进化成了仇怨，对祖国大陆的深深排斥，对昔日被殖民时期的莫名爱慕。随着这二十几年内地的赶超，香港已经完全没有任何昔日能够俯视内地的底气了。深圳的GDP已经超过香港大陆，一个是立足于祖国厚土，一个拘泥于港岛飞地，前者天生有着大国气象的延续，后者只有孤岛自残的自怨自艾。港村的衰退与癫狂，在于没有民族认同，没有文化自信，认贼作父，毫无廉耻，外不强而中烂。这种背弃自身民族和文化的结果，只有死路一条，不值得丝毫的同情和怜悯。

文化上同文同种的港村，都会疯癫如此。种族上如果大肆混血，会带来中华民族的昌盛富强么？答案是不能。所以，血肉的基因，更值得我们重视，我们主体民族的血脉必须纯洁。前文提到， 2017年以来，北京、上海、武汉、杭州、广州、成都等多个一二线城市，颁布了地方性移民法规，大幅放宽中国的绿卡政策。上海把这些年招来的质量参差不齐的外国留学生，列为国家急需、紧缺的高端外籍人才，可直接凭毕业文凭办理两年期居留许可，工作满三年的还可申请永久居留。？？？与此形成鲜明对比的，是上海对本国学生的落户政策，尤其对普通本科生，苛严之甚。上海作为中国最文明、最先进的大都市，都如此厚此薄彼的行事，其他地方可想而知。

为什么地方城市有权颁布地方性的移民法规？我们不是小米国的联邦制，没有地方法律各自独立的行为方式。我们必须要明确：中国人是炎黄子孙，中国人是中国的主人，在这片祖先留给我们的土地上，一切事情都必须中国人优先。资源是有限的，怎么能够一边对中国人计划生育，一边将我们的粮食、住房、教育等资源大量给洋垃圾使用呢？？

看看弗朗西斯他们家。5月19日，500多名非裔移民“突袭”戴高乐机场，抗议者拿着扩音器高喊：你们家不止属于你们自己、我们每个人都有权在你们家待着。再看看汉斯他们家，叙利亚难民怒斥汉斯们：不喜欢我们，你们可以滚出你们家。哥哥心里无数羊驼奔过，这都什么妖魔鬼怪？

回过头来，看看我们自己。中国今日的非法移民问题已经相当严重。网友调查显示：在广州的黑哥们已经形成了一定规模的聚集区，已达50万之多。其中相当一部分是非法滞留，他们还试图通过和中国人通婚获取在中国的定居地位，混血儿与日俱增。义乌定居的中东阿拉伯和南亚移民已经有数万人规模，他们建立了清真寺，严重影响周边居民生活。娶了中国妻子，无一例外的给她们戴上头巾。可以想象，如果外来非法移民的数量再上一个数量级，就基本没有管控的可能了。

法国曾经是欧洲举足轻重的大国，如今的人种构成彻底变了。曾经是100%的白人国家，现在却变成：非洲黑人+黑白混血儿占法国总人口的60%，纯种的白种人变成了35%，是少数稀有物种了。这种主体人种和民族的彻底转换，非洲黑人只用了不到80年的时间！如今的巴黎已经变成了臭气熏天、街道混乱、垃圾遍地、暴力犯罪和斗殴频发的城市。

网友说的好：民族利益是一个国家的最高利益，国破山河在，只要人还在，还有复国机会。但如果一个民族的种被灭了，这个民族就永远不能再起死回生。正视外国人用移民、非法居留、和中国女人生混血儿等方式对中国的和平入侵，这是一种理性。炎帝黄帝以来我们中国人就是东亚人民族，中国人和洋垃圾混血属于人种的退化，也许少数中国人和洋垃圾混血产生的问题不重要，但是如果大量中国人和洋垃圾混血将是中华民族的灾难，所谓的“国际交流”也不应该是这样的。

不要跟哥哥掰扯白左圣母婊那套。小米奴役黑哥们快两百年，欧诸还寻思着做回封建领主，套住黑绿给他们当仆人。结果，应了那句老话，天道好轮回,苍天饶过谁？现在他们徘徊在民族被稀释灭种的边缘。他们欠黑哥们的，心术不正，活该遭反噬。我们中国人民不一样，我们在自己的土地上，本本分分，艰辛苦难几千年，创建自己的家园，延续自己的文明。我们欠黑哥们what？

我们自己战天斗地，40年改革开放蒸蒸日上的今天，为什么要引乱入室，混杂我们的民族基因？我们必须自强自立，保证民族的延续，否则，真正的摸得着看得见的颜色革命，就近在你我的眼前了。这颜色可能是黑、是绿、是白、是褐，但不再是黄了。

 

这篇裹脚文，由几颗烂瓜，引出些许牢骚。梦之呓语，当真是否，全在自心。

最后想说，中华民族的基因是世界上最优秀的基因，请你珍视珍惜珍重。每一个中国人，如果你爱这个民族，这片土地，你有祖先的坚韧，你有披靡的勇气，就再一次聆听太祖手持红缨之日的挥斥方遒：世界，我们的世界；国家，我们的国家；社会，我们的社会。我们不说，谁说？我们不干，谁干？

谨以此文与大家共勉。

## 贸易战再起波澜
就在第十一轮中美经贸磋商之际美国悍然宣布加征关税，而在今晚美股开盘之前也公布了对美加税的反制措施，一直看似缓和的国际局势陡然升温，在很多朋友看来可能全面摊牌即将到来。

在我看来现阶段灯塔还赌不起全面摊牌。

为何如此笃定灯塔现阶段还赌不起全面摊牌？原因有这么几点，
* 首先对于美方本次加税细节可以看出，灯塔还不敢马上全面加征关税。就本次加税美国贸易代表发布的关税实施方案显示，对于5月10日前出口、6月1日前进入美国的中国商品仍然加征10%的关税，而不适用于加征25%的关税。其中的意味显而易见，未来两周左右特大嘴随时变脸。
* 其次，这一点在之前反复讲过，就是在彻底解决北极熊之前灯塔无法接受全面摊牌的后果。
* 最后，就是本次特大嘴变脸的重要原因就是灯塔威逼利诱老大哥做出对重要伙伴的重要原则立场的选择，但是老大哥软硬不吃！同时，这时老大哥一旦在重要原则立场上有所让步，原本已经岌岌可危的重要伙伴怕是将面临快速覆灭。同时另一重要战略伙伴关系也将被彻底离间。形势如此，我们也必当如此！

 对于贸易战以及将来几乎不可避免的金融战，我曾经提到过很多武器，这次我们一如既往的稍稍用了汇率，最近几天的人民币的走势想必大家也一目了然。
 
 既然你要加税，那来而不往非礼也，于是今晚我们礼貌性的在美股开盘之前宣布对美加征关税。想必此夜大洋彼岸的股票市场和期货市场正经历一场血雨腥风。也许大家也说了，这段时间我们的市场也是血雨腥风，我想问问这些朋友，难道你没有收到我一月之前推送的信息，我看了一下手机，但是推送信息的时间是4月4日早上9点30多，当时给出的结论是：未来两年内上证下行一千点可期，当时的指数是三千二百多点，而此时仅仅过去一月有余，指数到了什么地步身在市场的你想必最清楚。

# 2019-02-26牛市来了？朋友你想多了

原创 大文豪曼因斯坦  大文豪曼因斯坦  2019-02-26  
自步入2019以来股市的躁动让很多朋友心绪难平，尤其昨日股市出现夺命长阳之后很多读者朋友纷纷留言问及股市。但凡问这个问题的人都没有认证看我在元旦发出的文章《2019，你无需悲观》，在这篇文章中我给出很多明示或暗示，当然也有不少朋友已然领悟其中很多暗示。就暗示的内容我曾经想过很多次是否需要明示给大家，考虑再三还是维持当初的表述。在这里我再重申一遍，2019是比较乐观的一年，但却是投机之年，绝非慢慢长牛启动之年。
 
当初《2019，你无需悲观》中这段文字你需要重温一下：“对于股市，好消息是随着注册制的试点和落地赌场又向印钞厂迈进了一步。本次会议还是单独提及了资本市场的话题“资本市场在金融运行中具有牵一发而动全身的作用，要通过深化改革，打造一个规范、透明、开放、有活力、有韧性的资本市场，提高上市公司质量，完善交易制度，引导更多中长期资金进入，推动在上交所设立科创板并试点注册制尽快落地。”关于18年的股市去年11月初就迫不及待的发文提醒过大家，并且年中反复提醒，当初看似疯狂的断言可惜也是一语成谶。伴随着一年时间近千点的下跌，目前的A股已经完全处于可控区域，结合对于19年整个经济政策的定调和纲领，19年的A股将迎来投机之年。如果说17年底我对18年A股画的是一根大阴线的话，那么今年初我为19年画的是一根带下影线的阳线。但有一点必须要说明这是一个假阳。
积极的财政政策和定向宽松的货币政策为19年的市场提供了乐观预期的基础和相对宽松一点的资金环境，市场向乐观的方向发展也是理所当然。另外，18年整体的大跌虽然重心下移到理想的点位，但一定程度上还是影响了股市直接融资的功能，而在今年直接融资这一功能需要进一步发挥，其中也包括鼓励并购重组支持科技及新经济行业。最后一点就牵涉到注册制了，如果注册制一旦全面展开你们可以想象一下有多少公司的股票将变成废纸，从时间上看注册制从试点到落地时间不会太久，而19年的宏观政策环境虽然不算太好，但在整个全球经济下行的周期中已实属不易，所以不难想象这一年将是很多资本在他们手里的股票真正变成废纸之前的出逃之年。综上种种不难得出2019将是A股的投机之年。”
 
没有明示这样的机会主要是考虑投机的机会并非大多数人可以把握。对于是否是投机之年也许事后你们才能发现，但我可以推断大概率如此。既然是投机相信大多数人是难以全身而退的，为何你觉得你会成为其中的幸运者。我为19年的行情画了一根带下影阳线，而且明确说这是假阳。至于是不是阳线你们年底便知，但至于是不是假阳你们明年才能知晓。

# 2019-01-31微信好友“含文率”知多少

原创 大文豪曼因斯坦  大文豪曼因斯坦  2019-01-31  
年终岁尾之际这篇文章就不谈局势经济之类严肃的话题了，正好有个轻松的话题可以跟大家互动一下。

 关于推荐的书简单说说，也推荐几本。
1.经济和金融书：曼昆的《经济学原理》，深入浅出，结合案例，比较浅显易懂。《两次全球大危机的比较研究》应该是宏观层面解读的，通过目前的实操手段去看表面上理解应该还可以。《稳定不稳定的经济》和《中央银行与货币供给》专业性挺强，还是需要好好啃一下，可以先看下《从零开始度懂金融学》先打下底，对金融学有个初步了解。《政治图鉴》和《资源战争》应该有助于理解目前全球政治经济行为，不过还没来得及读。
2.哲学书：老子《道德经》其实也只是是哲学流派思想。哲学是对于人生的有系统的反思的思想，有志的人应该尽量有自己处事的哲学思想，这些会在关键时刻影响决策和处事方式。《道德经》挺难懂得，读之前可以先看下《中国哲学简史》和《哲学简史》。
