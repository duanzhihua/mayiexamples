# 课程概述

人工智能、无人驾驶技术等新兴科技被视为人类继蒸汽机时代、电器时代和信息时代之后的第四次工业革命。

中国也是目前本次工业革命的重要试验场。科技的发展极大地改进了保险生态环境，改良保险行业痛点，改变保险的传统使命。

本课程系统地讲解未来影响中国保险行业发展的十大重要科技，这10项科技包括区块链技术、人工智能、物联网、云计算、大数据、车联网、无人驾驶汽车、无人机、基因检测、可穿戴设备，每项技术均阐明技术原理，分析保险行业相关痛点和运用相关科技的解决思路，并且讨论运用场景以及技术结合过程的机遇、挑战。

在技术分析地基础上，进一步讲解保险科技发展的创新渠道、细分领域的创新，包括产品创新、保险营销、企业运营、信息咨询等多个方面。

课程还将辨正保险科技认识的常见误区、展望保险科技的发展前景与生态环境。

通过本课程的学习，学生将系统的了解到创新科技在保险领域的具体运用，思考现有保险行业痛点下如何进行赋能式或颠覆式的保险科技创新，以及保险科技与其他金融相关领域的融合、对整体保险行业的冲击和面临的机会与挑战。

## 3.2 人工智能保险行业痛点
![](imgs/ai01.png)










那么跟人工智能所对应的保险行业，它的行业痛点是什么？

保险行业的第一个行业痛点是
### 保险产品的营销。
在目前的保险行业里面，主要的分销渠道依然是人，个人代理和银行销售。庞大的代理人的团队造成了保险公司非常多的问题。

第一个问题，保险代理人从业门槛较低，人员流动比较高，从而导致了专业性难以保障等等问题，这也是导致了很多消费者投诉的一个问题，成本居高不下，投保人支付高昂的保费，最后并没有得到良好的客户体验。另外由于佣金制度的存在，代理人在服务的过程里面往往会以促成交易为导向，所以倾向于给消费者推销他能拿到更高佣金的这一部分产品，导致销售不恰当或者是不合适，降低了保险保障的效率。另外在佣金抽成期过后，消费者明显的感觉到服务质量是在下降，这些问题共同造成了保险人的品牌的忠诚度非常低，而且退保率也会节节的上升。

保险业的另外一个痛点是
### 保险产品费率和服务范围。
我们都知道传统的保险产品都是根据大数法则，它主要依据的是历史的数据和精算的模型来进行公平的保费计算，对于风险标的的交互十分有限，也没有办法掌控到个人的风险水平，因此就有现实中许多更高风险的人才会愿意购买保险，也就是我们经常强调的逆向选择的问题。

同时由于对风险水平不同的个体以相同的保险费率来收取保费，那么这必定会侵蚀到那些低风险人群他的保险意愿和他的保险利益，由此还导致了道德风险的问题，也就是说在购买了保险以后可能会疏于对风险的防范。传统的保险主要是服务于头部人群，也就是高风险人群的需求，由于技术成本的限制，难以满足长尾人群，也就是那些低风险人群的需求，没有办法提供深度定制的个性化的保险服务来满足投保人的保险需求，大大的限制了传统保险产品能够覆盖的范围。另外面对着复杂的保险需求，由于保险经纪人个人能力的限制，它没有办法给投保人推举出非常合适的保险组合和保险产品。

保险公司的另外一个痛点还存在
### 核保和承保的范围。
通常来说投保人购买了一份保险，是要经过填写保单、交保费、保单的扫描、核保签发保单等等诸多的环节，不仅要求投保人到柜台办理，而且这个环节非常复杂，时间也比较冗长，过程不够透明，投保人难以直接的了解投保的进度。而对于一些频次高额度小碎片化的这些保险产品或者是标准化的网络保险，采用人工核保的方式，不仅成本高而且效率低，除此以外，打印各种单据不仅不容易保存也不够低碳和环保。

保险业的另外一个痛点体现在
### 保险产品的理赔和反欺诈，
投保人利用信息不对称进行骗保，或者与第三方的机构中介，或者是医疗机构合谋虚假理赔，或者是提高理赔的金额，保险赔偿金被冒领等等，许多问题一直都是困扰着保险行业发展的问题，使得保险公司不得不在核赔上加大人工投入来减少保险欺诈所带来的损失。而对于消费者而言，复杂的理赔手续导致的低效率以及对保险条款的理解不同，或者是核赔人员的主观判断失误，导致了赔偿金额低或者是拒赔，会加深消费者对保险机构的不信任，破坏客户关系，最后会导致客户的流失

## 3.3 人工智能解决思路
保险行业所存在的痛点可以很好地借助人工技术，人工智能的发展来进行解决。比如说保险公司可以建立智能机器人服务系统，从而实现7×24小时，也就是每周七天无间歇的工作的在线问答服务。人工智能将为客户提供保单查询条款解读保费计算等等一站式的服务，客户可以通过像智能机器人提问，智能机器人以与人类沟通的这样一种方式来解答客户的问题，并且通过自身不断的学习的优化改善问题的识别率，提高回答问题的准确率。建立在人工智能基础上的智能保护系统可以很好地解决保险营销的困扰智能保护或者叫智能投顾，英文叫RobotAdviser，他的系统不仅可以大大的降低保险公司的人力成本，而且可以避免佣金制度所带来的弊端，同时可以针对不同客户提供个性化私人定制服务，实现精准营销。

目前私人定制的保险经纪服务主要针对的是高端的客户，而智能保顾系统也可以使得广大普通的老百姓也可以体会到个性化的保险服务，实现我们国家所提倡的普惠金融。客户可以在线填写保单系统，向系统提供家庭的情况，收入的情况等等诸多的相关的信息，然后机器人通过一系列的算法来分析，在不同的阶段它的保险需求，并且结合他们的保险需求和产品进行匹配，为消费者提供一套量身定做的保险解决方案，使得客户可以感受到保险对自身效用的提高，这种良好的客户体验可以提高消费者的忠诚度，进一步的降低消费者的退保率。

保险公司可以利用人工智能建立一套无纸化的在线人工智能核保和承保体系，这样的体系可以解决大部分保险产品在核保和承保的问题，少数不能解决的疑难问题，再转由人工进行处理。如果能够按照这种模式进行运作的话，保险公司的运营成本将会大大的降低，效率会得到大量地提高。人工智能系统可以通过积累人工核保的经验案例，深度学习各种核保知识，通过人机交互的方式与客户进行在线交流，了解客户的信息和风险特征，最终出具是否承保的决定。

在这样的一个过程里面，利用人脸识别等生物健全识别技术对投保人的身份进行识别确认，在线自助填单，并且上传各种单据，通过OCR也就是OpticalCharacterRecognition识别技术进行处理，核保通过了以后，投保人可以在线缴费，可能最快几分钟以内就可以进行现场的承保，大大的节省了投保人的时间成本，而且整个过程可以实现无纸化低碳环保，同时避免了保单毁损和遗失等等的风险。通过人工智能的手段，保险公司可以实现和保险标的之间的实时交互，改变以往的保险定价模式。
### 动态费率
通过对于实时数据的收集和费率因子的管控，基于动态模型评估个体风险水平，并且实施厘定保险费率。通过建立一套奖惩机制，比如说对于那些能够达到保险公司要求的，能够进行良好的自我约束的，实施自我风险管理的投保人，降低他们的保费，而对于那些存在道德风险或者是不能达到要求的这些客户，可以适当的提高他们的保费，借此来控制逆向选择和道德风险的问题，帮助保险在防灾防损方面发挥更大的作用。人工智能的运用还大大的扩大了保险的服务范围，基于云计算、大数据等技术之上的人工智能系统运算速度和规模都将得到巨大的提升。

得益于这样的提升，在处理保险业务的效率上，保险公司将可以大大的提高。在过去依靠人力无法承保的项目，比如说退货运输险，在互联网时代，我们今天看到是一个非常成功的一个典范。而基于大数据测算的智能定价模型，可以根据不同标的的历史数据进行不同的费率厘定，可以将不可保的风险转化为可保的风险，并且能够满足这些长尾客户，也就是我们讲的这种低风险人群的个性化需求，从而大大的扩大了保险的服务范围，使得保险能够在更大的领域里面对社会经济发挥更大的作用。保险公司通过智能合约来履行保险合同，并且处理理赔，一方面减少了人力成本，另一方面也降低了人工处理可能带来的瑕疵和纰漏。

通过图片识别技术，将投保人提供的理赔图像在互联网中进行比对，可以减少信息不对称所导致的骗保。通过对于医疗记录的甄别，并且利用大数据技术进行对比测算，可以减少合谋带来的保险欺诈。通过人脸识别可以避免保险金的冒领等情况。通过人工智能系统的识别，通过人工智能系统的处理，大大的提高了保险公司的运行效率，减少运营成本，并且有助于维护保险机构和投保人之间一种良性的客户关系。
